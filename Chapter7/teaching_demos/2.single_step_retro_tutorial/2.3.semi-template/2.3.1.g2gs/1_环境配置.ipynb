## 这份环境为什么和原始 `torchdrug` 教程不同？

原始教程依赖完整的 `torchdrug` 框架，以及一些为高性能图计算准备的附加库。为了让同学们更容易聚焦到 **G2Gs 的核心思想**，本教学版做了两点简化：

- 只保留理解 G2Gs 必需的基础依赖。
- 把底层逻辑拆成可读、可调试的教学代码。

因此，你可以把这里的环境理解为：**一个专门服务于教学讲解的轻量环境**。


## 推荐版本说明

下面这套版本是为本教材整理过的一组兼容配置。你不需要死记每个版本号，只需要知道两件事：

1. 这些版本已经覆盖了本教程需要的功能。
2. 如果你后面遇到环境问题，优先检查自己安装的版本是否和这里差别过大。

| 组件 | 推荐版本 | 作用 |
|------|----------|------|
| Python | `3.13` | 运行 notebook 与教学代码 |
| PyTorch | `2.9.0` | 张量计算与教学模型 |
| RDKit | `2025.9.5` | 分子解析、处理和绘图 |

| matplotlib | `3.10.8` | 基础可视化 |
| networkx | `3.6.1` | 图结构辅助分析 |
| ipykernel | `7.1.0` | 注册 notebook kernel |
| jupyterlab | `4.5.4` | 本地交互开发（可选） |

如果你只是跟着教材学习，按下面的安装命令执行即可，不需要自行调整版本。


```python
import os
import sys
from pathlib import Path


def find_project_root(start=None):
    here = Path(start or os.getcwd()).resolve()
    if here.is_file():
        here = here.parent
    for candidate in [here, *here.parents]:
        if (candidate / "teaching_demos").exists() and (candidate / "source_repos").exists():
            return candidate
    raise FileNotFoundError("无法定位项目根目录")


PROJECT_ROOT = find_project_root()
TUTORIAL_DIR = PROJECT_ROOT / "teaching_demos/2.single_step_retro_tutorial/2.3.semi-template/2.3.1.g2gs"
CODE_DIR = TUTORIAL_DIR / "code"
DATA_DIR = TUTORIAL_DIR / "data"
PROCESSED_DIR = DATA_DIR / "processed"
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

if str(CODE_DIR) not in sys.path:
    sys.path.insert(0, str(CODE_DIR))

print(f"项目根目录: {PROJECT_ROOT}")
print(f"教程目录: {TUTORIAL_DIR}")
print(f"代码目录: {CODE_DIR}")
print(f"数据目录: {DATA_DIR}")

ENV_DIR = PROJECT_ROOT / "envs" / "g2gs_tutorial_envs"
print(f"环境目录: {ENV_DIR}")
```

## 1. 创建环境并安装依赖

这一部分会在项目的 `envs/` 目录下新建一个独立环境，名字是 `g2gs_tutorial_envs`。

这样做的好处是：

- 不会污染你系统里已有的 Python 环境。
- 不会和 GLN、LocalRetro 等其他教学环境冲突。
- 如果后面环境损坏，直接删除并重建就可以。

建议同学们第一次运行时耐心等待，这一步通常是整个教程里最耗时的一步。


```python
%%bash -s "$PROJECT_ROOT"
PROJECT_ROOT="$1"
ENV_DIR="$PROJECT_ROOT/envs/g2gs_tutorial_envs"

source "$(conda info --base)/etc/profile.d/conda.sh"

if [ ! -d "$ENV_DIR" ]; then
    echo "==> 创建 Conda 环境: $ENV_DIR"
    conda create -y -p "$ENV_DIR" python=3.13
else
    echo "==> 环境已存在: $ENV_DIR"
fi

conda activate "$ENV_DIR"
python -m pip install --upgrade pip

pip install --index-url https://download.pytorch.org/whl/cpu torch==2.9.0
pip install                 rdkit==2025.9.5                 pandas==3.0.0                 matplotlib==3.10.8                 networkx==3.6.1                 ipykernel==7.1.0                 jupyterlab==4.5.4 jupyter notebook
```

## 2. 注册 notebook kernel

如果你在 JupyterLab 或 VS Code Notebook 中学习，建议把这个环境注册成单独的 kernel。

这样后面打开 notebook 时，你只需要选择 `G2Gs Tutorial`，就能确保代码运行在正确环境里，不容易出现“明明装了库却导入失败”的问题。


```python
%%bash -s "$PROJECT_ROOT"
PROJECT_ROOT="$1"
ENV_DIR="$PROJECT_ROOT/envs/g2gs_tutorial_envs"

source "$(conda info --base)/etc/profile.d/conda.sh"
conda activate "$ENV_DIR"

python -m ipykernel install --user --name g2gs_tutorial --display-name "G2Gs Tutorial"
echo "==> Kernel 已注册: G2Gs Tutorial"
```
